# Pymaceuticals Inc.
---### Analysis
####In this study, 249 mice identified with Suqamous Cell Carcinoma tumor growth. They recieved various pharmacuetical treatment. Over the course of 45 days, tumor development was observed and measured. The purpose of this study was to compare the performance of Pymaceuticals' drug of interest, Capomulin, versus the other formulas.

```python
import pandas as pd

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Merge the two datasets on the common column "Mouse ID"
combined_data = pd.merge(mouse_metadata, study_results, on="Mouse ID")

# Calculate the summary statistics for each drug regimen
summary_stats = combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(
    Mean="mean",
    Median="median",
    Variance="var",
    Standard_Deviation="std",
    SEM="sem"
).reset_index()

# Display the summary statistics table
print(summary_stats)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
mouse_metadata = pd.read_csv("data/Mouse_metadata.csv")
study_results = pd.read_csv("data/Study_results.csv")

# Merge the dataframes if necessary
merged_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")

# Perform analysis
mean_tumor = merged_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
median_tumor = merged_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
variance_tumor = merged_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
std_tumor = merged_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
sem_tumor = merged_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()

# Create summary DataFrame
summary_data = pd.DataFrame({
    "Mean of Tumor": mean_tumor,
    "Median of Tumor": median_tumor,
    "Variance of Tumor": variance_tumor,
    "Std of Tumor": std_tumor,
    "SEM of Tumor": sem_tumor
})

# Display the summary statistics table
print(summary_data)

# Optionally, display the summary statistics table in a more visual way
plt.figure(figsize=(17, 25))
summary_data.plot(kind='bar')
plt.title('Summary Statistics of Tumor Volume by Drug Regimen')
plt.xlabel('Drug Regimen')
plt.ylabel('Tumor Volume')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Quartiles, Outliers and Boxplots

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Merge the dataframes
merged_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")

# Group the data by 'Drug Regimen'
grouped_data = merged_df.groupby('Drug Regimen')

# Create a list to hold tumor volume data for each regimen
tumor_volumes = [grouped_data.get_group(regimen)['Tumor Volume (mm3)'] for regimen in grouped_data.groups]

# Create the boxplot
plt.figure(figsize=(12, 8))
plt.boxplot(tumor_volumes, labels=grouped_data.groups.keys())
plt.title('Tumor Volume by Drug Regimen')
plt.xlabel('Drug Regimen')
plt.ylabel('Tumor Volume (mm3)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Identify outliers
for regimen, volumes in grouped_data['Tumor Volume (mm3)']:
    quartiles = volumes.quantile([0.25, 0.5, 0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr)
    outliers = volumes[(volumes < lower_bound) | (volumes > upper_bound)]
    print(f"Outliers for {regimen}: {outliers}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Merge the dataframes
merged_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")

# Filter the data for the Capomulin regimen
capomulin_data = merged_df[merged_df['Drug Regimen'] == 'Capomulin']

# Group the data by mouse ID and calculate the average tumor volume for each mouse
average_tumor_volume = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
mouse_weight = capomulin_data.groupby('Mouse ID')['Weight (g)'].mean()

# Combine the data into a single DataFrame
capomulin_summary = pd.DataFrame({
    'Mouse Weight (g)': mouse_weight,
    'Average Tumor Volume (mm3)': average_tumor_volume
})

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(capomulin_summary['Mouse Weight (g)'], capomulin_summary['Average Tumor Volume (mm3)'], color='blue')
plt.title('Mouse Weight vs. Average Tumor Volume for Capomulin Regimen')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.grid(True)
plt.show()
```

## Correlation and Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Load the data
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Merge the dataframes
merged_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")

# Filter the data for the Capomulin regimen
capomulin_data = merged_df[merged_df['Drug Regimen'] == 'Capomulin']

# Group the data by mouse ID and calculate the average tumor volume for each mouse
average_tumor_volume = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
mouse_weight = capomulin_data.groupby('Mouse ID')['Weight (g)'].mean()

# Combine the data into a single DataFrame
capomulin_summary = pd.DataFrame({
    'Mouse Weight (g)': mouse_weight,
    'Average Tumor Volume (mm3)': average_tumor_volume
})

# Calculate the correlation coefficient
correlation_coefficient = capomulin_summary.corr().iloc[0, 1]
print(f"Correlation coefficient: {correlation_coefficient}")

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(capomulin_summary['Mouse Weight (g)'], capomulin_summary['Average Tumor Volume (mm3)'])

# Print the linear regression results
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-squared: {r_value**2}")
print(f"P-value: {p_value}")
print(f"Standard error: {std_err}")

# Plot the scatter plot with the regression line
plt.figure(figsize=(10, 6))
plt.scatter(capomulin_summary['Mouse Weight (g)'], capomulin_summary['Average Tumor Volume (mm3)'], color='blue', label='Data Points')
plt.plot(capomulin_summary['Mouse Weight (g)'], slope * capomulin_summary['Mouse Weight (g)'] + intercept, color='red', label='Regression Line')
plt.title('Mouse Weight vs. Average Tumor Volume for Capomulin Regimen')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.legend()
plt.grid(True)
plt.show()
```
